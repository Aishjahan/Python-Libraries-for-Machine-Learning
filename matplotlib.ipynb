# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x = ["python","c","c++","java"]
y = [85,78,60,82]
z = [20,30,40,50]
plt.xlabel("language",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.title("My data",fontsize=30)
c = ["r","y","b","m"]
plt.bar(x,y,width=0.4,color="g",align="edge",edgecolor="r",linewidth=2,linestyle=":",alpha=0.6,label="My data")
plt.legend()
plt.show()
```

```python
import numpy as np

x = ["python","c","c++","java"]
y = [85,78,60,82]
z = [20,30,40,50]

width=0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.title("My data",fontsize=30)

plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity2")

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
x = ["python","c","c++","java"]
y = [85,78,60,82]
z = [20,30,40,50]

plt.xlabel("language",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.title("My data",fontsize=30)

plt.barh(x,y,color="g",align="edge",label="My data")
plt.legend(loc="best")
plt.show()
```

# Scatter Plot

```python
x=[1,2,3,4,5,6,7,3,4,5,1,2,3,4,7,2,3,5,7,9,3,4,8,9]
y=[2,3,1,4,5,3,6,4,5,6,7,1,2,5,8,9,3,5,7,2,5,7,2,5]
plt.scatter(x,y,color="r",s=400)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
sizes=[200,500,300,700,100,300,800]
plt.scatter(x,y,color="r",s=sizes,alpha=0.7,marker="*",edgecolor="y")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

# Pie Chart

```python
x = [10,20,30,40]
y = ["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
x = [10,20,30,40]
y = ["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,labeldistance=1.2,startangle=90)
plt.title("Python Data")
plt.legend(loc=4)
plt.show()
```

# Stem Plot

```python
x = [1,4,3,2,5,7,8]
y = [4,6,1,5,9,3,6]

plt.stem(x,y,label="python")
plt.legend()
plt.show()
```

```python
x = [1,4,3,2,5,7,8]
y = [4,6,1,5,9,3,6]

plt.stem(x,y,linefmt=":",markerfmt="ro")
plt.show()
```

# Box Plot

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,widths=0.2,labels=["python"],patch_artist=True,showmeans=True)

plt.legend()
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]
x1 = [80,40,60,34,56,79,42,57]

y=[x,x1]
plt.boxplot(y,widths=0.2,labels=["python","java"],patch_artist=True,showmeans=True)

plt.legend()
plt.show()
```
