# BAR PLOT

```python
res = sns.barplot(x ='cyl',y ='mpg', data = mtcars)
plt.show()
```

```python
res = sns.barplot(x ='cyl',y ='mpg', data = mtcars, color='yellow')
plt.show()
```

```python
res = sns.barplot(x ='am',y ='mpg', data = mtcars, palette = "Set2")
plt.show()
```

```python
res = sns.barplot(x ='gear',y ='hp', data = mtcars, palette = "rocket")
plt.show()
```

# COUNT PLOT

```python
sns.countplot(x='cyl', data = mtcars, palette = "Set1")
```

```python
sns.countplot(x='carb', data=mtcars, hue='carb', palette='Set2', legend=False)
plt.show()
```

```python
sns.countplot(y='gear', data=mtcars, hue='gear', palette='rocket', legend=False)
plt.show()
```

```python
sns.countplot(x = 'gear', hue = 'cyl', data = mtcars, palette = "Set1")
```

# DISTPLOT

```python
sns.distplot(mtcars.mpg, bins=10, color='g')
```

# HeatMap

```python
# Select only numeric columns
numeric_mtcars = mtcars.select_dtypes(include=[np.number])

# Generate the heatmap
sns.heatmap(numeric_mtcars.corr(), cbar=True, linewidth=0.5)
plt.show()
```

# SCATTER PLOT

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue='species', size='species')
```

# PAIR PLOT

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species', palette = "Set1")
```

# LMPLOT

```python
sns.lmplot(x='petal_length', y='petal_width',data=iris)
```

```python
sns.lmplot(x = 'petal_length', y='petal_width', hue='species', data=iris, markers=["o","*","^"])
```
